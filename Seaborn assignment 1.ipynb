```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data=sns.load_dataset('titanic')
```

```python
data.head()
```

```python
data['age'].fillna(data['age'].mean(),inplace=True)
```

# Problem 1 : 
Plot a Heatmap on the correlation between multiple columns

```python
x=data.corr()
```

```python
sns.heatmap(x)
```

# Problem 2 : 
Plot the distribution of age column using a Boxplot on the basis of sex column

```python
sns.catplot(x='sex',y='age',kind='box',data=data)
```

# Problem 3 : 
Plot the distribution of age using a ViolinPlot on the basis of both Pclass and Sex column

```python
sns.catplot(x='pclass',y='age',kind='violin',hue='sex',split=True,data=data)
```

# Problem 4 : 
Plot the count of Survivors and Non-Survivors

```python
sns.catplot(x='survived',kind='count',data=data)
```

# Problem 5: 
Plot the count of survivors as well as non-survivors on the basis of Pclass

```python
sns.catplot(x='survived',hue='pclass',kind='count',data=data)
```

# Problem 6: 
Plot a scatterplot between age and fare column on the basis of both Pclass and Sex column

```python
sns.relplot(x='age',y='fare',hue='pclass',style='sex',data=data)
```
